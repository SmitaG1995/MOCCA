# The calibration factors

The data analysis is done using a MATLAB package called ANACONDA 2 (https://github.com/gasfas/ANACONDA_2). This notebook calculates the calibration factors that are used as input for the ANACONDA 2 package.

```python
v_mb =slopex*1e6
dx = interceptx
dy = intercepty
```

# Use the calibration factors in raw data

```python
data['x'] , data['y'] = rot_axes(data['x'],data['y'],theta) #the detector image is rotated
data['x'] = data['x'] - dx #we translate the image
data['y'] = data['y'] - dy #we translate the image

```

```python
def cart2sper(data):
    data['phi'] = np.arctan2(data['dpy'], data['dpx'])
    data['r'] = sqrt(data['dpx']**2 + data['dpy']**2 + data['dpz']**2)
    data['theta'] = np.arccos(data['dpz'] / data['r'])
    return data
```

# Fine calibration of dx dy

```python
data
```

```python
def f(dx_new= -0.6, dy_new = -0.3, v1 = 400, tshift_new = 0.5):
    global data
    data['x'] = data['x'] - dx_new #we translate the image
    data['y'] = data['y'] - dy_new #we translate the image
    ef = electric_field(v1,v2,dist)
    global tof_no_p, tshift,factor
    tof_no_p = tof_0(m,tshift + tshift_new,factor)
    map_dict2 = { m[i] : tof_no_p[i] for i in range(len(m)) }
    data['tof_no_p'] = data.m2q.map(map_dict2)
    
    ##plotting
    data = p0(data, m, c, tof_no_p, v_mb, ef)
    data = dp(data, ef)
    data = cart2sper(data)
    momentum_plot2(data, 'dpx','dp_norm','dpy','dp_norm','dpz', 'dp_norm', v1, 28) #to find v1 dx dy
    momentum_sphere(data, 'dp_norm','r','dp_norm','theta','dp_norm', 'phi', v1, 28) #to find v1 dx dy
#     momentum_plot(data, 'dpx','dpy','dpz','dpx', v1, 12) to find tshift
    ##after plotting
    data['x'] = data['x'] + dx_new #we translate the image back to original
    data['y'] = data['y'] + dy_new #we translate the image back to original
    return dx_new, dy_new, v1, tshift_new 

fine_cal = interactive(f,dx_new=(-3,2,0.1),dy_new=(-2,2,0.2),v1=(300,500,10),tshift_new=(-5,5,0.5))
display(fine_cal)
```

```python
dx_new, dy_new, v1, tshift_new  = fine_cal.result
```

# The final calibration factors

```python
print('The calibrated theta value (degrees) is',theta)
print('The molecular beam velocity (m/s) is', v_mb)
print('The calibration factor dx=', interceptx + dx_new)
print('The calibration factor dy=', intercepty + dy_new)
print('The TOF calibration factor to=', tshift + tshift_new)
print('The voltage grid (v1)=',v1)
```
